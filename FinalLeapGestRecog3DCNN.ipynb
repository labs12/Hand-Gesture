### Pre-Processing for train data

```python
ls_path = os.path.join('./dataset/leapGestRecog/train_data/train/1/')
listing = os.listdir(ls_path)

frames = []
img_depth = 0

for img in tqdm(range(300)):
    for imgs in listing:
        if img_depth <30:
            img = os.path.join(ls_path,imgs)
           # print(img)
            frame = cv2.imread(img)
            frame = cv2.resize(frame,(img_rows,img_cols),interpolation=cv2.INTER_AREA)
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frames.append(gray)
            img_depth = img_depth+1
    
    #else:
        #break
    input_img = np.array(frames)
    ipt = np.rollaxis(np.rollaxis(input_img,2,0),2,0)
    ipt = np.rollaxis(ipt,2,0)
    X_tr.append(ipt)
        
    
print(ipt.shape)
num_samples = len(X_tr)
print(num_samples)
```

```python
X_tr_array = np.array(X_tr) #convert the frames read to array
num_samples = len(X_tr_array)
print('Total training samples:',num_samples)
X_val_array = np.array(X_val)
num_val_samples = len(X_val_array)
print('Total validation samples:', num_val_samples)
```

```python
label_train = np.ones((num_samples),dtype=int)
label_train[0:300] = 0
label_train[300:600] = 1
label_train[600:900] = 2
label_val = np.ones((num_val_samples),dtype=int)
label_val[0:100] = 0
label_val[100:200] = 1
label_val[200:300] = 2
```

```python
train_data = [X_tr_array,label_train]
```

```python
(X_train,y_train) = (train_data[0],train_data[1])
print('Training data shape:',X_train.shape)
```

```python
train_set = np.zeros((900,30,64,64,3),'float32')
```

```python
for h in range(300):
    train_set[h][:][:][:][:]=X_train[h,:,:,:]
```

```python
val_data = [X_val_array,label_val]
```

```python
(X_val,y_val) = (val_data[0],val_data[1])
print('Validtion data shape:', X_val.shape)
```

```python
val_set = np.zeros((300,30,64,64,3),'float32')
```

```python
for h in range(300):
    val_set[h][:][:][:] = X_val[h,:,:,:]
```

```python
nb_classes = 3
Y_train = to_categorical(y_train,nb_classes)
```

```python
Y_val = to_categorical(y_val,nb_classes)
```

```python
train_set = train_set.astype('float32')
train_set-=np.mean(train_set)
print("Mean Train:",train_set)
train_set /=np.max(train_set)
print("Max Train", train_set)
```

```python
val_set = val_set.astype('float32')
print(np.mean(val_set))
val_set -=np.mean(val_set)
print(np.max(val_set))
val_set /=np.max(val_set)
```

```python
def build_model():
    model = Sequential()

    model.add(Conv3D(8, (3,3,3), input_shape=(frame_size,img_cols,img_rows,3),activation='relu'))
    model.add(Conv3D(16, (3,3,3),activation='relu'))
    # model.add(BatchNormalization())
    model.add(MaxPool3D())
    # model.add(Dropout(0.25))

    model.add(Conv3D(32, (3,3,3),activation='relu'))
    model.add(Conv3D(64, (3,3,3),activation='relu'))
    model.add(BatchNormalization())
    model.add(MaxPool3D())
    model.add(Dropout(0.25))

    model.add(Flatten())

    model.add(Dense(4096, activation='relu'))
    model.add(Dropout(0.5))

    model.add(Dense(1024, activation='relu'))
    model.add(Dropout(0.5))

    model.add(Dense(nb_classes, activation='softmax'))

    return model
    
    
```

```python
model = build_model()
```

```python
model.summary()
```

```python
save_dir = os.path.join(os.getcwd(),'saved_model')
print(os.getcwd())
model_name = 'LeapGesRecog'
if not os.path.isdir(save_dir):
    os.makedirs(save_dir)
model_path = os.path.join(save_dir,model_name)
checkpoint = ModelCheckpoint(model_path,monitor='val_acc',save_best_only=True,verbose=1)
earlystop = EarlyStopping(monitor='val_loss',patience=50,verbose=1)
```

```python
model.compile(optimizer ='RMSprop',
              loss = 'categorical_crossentropy',
                metrics = ['acc'])
```

```python
batch_size = 32
epoch = 50
lr_reducer = ReduceLROnPlateau(monitor='val_loss',factor=0.05,
                              cooldowm=0,patience=10,min_lr=0.005/(2^4),verbose=1)
```

```python
history = model.fit(train_set,Y_train,
                  validation_data=(val_set,Y_val),
                  batch_size=batch_size,
                  epochs=epoch,
                  callbacks=[checkpoint,lr_reducer])
```
